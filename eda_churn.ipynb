## Merging dataframes

#### Measuring dataframes

```python
len(df_contracts)
```

```python
len(df_customers)
```

```python
len(df_services)
```

```python
# Rename columns
df_services.rename(columns={'customerID': 'ClientID'}, inplace=True)
```

```python
# Merging customers and services
df_temp = df_customers.merge(df_services, on=['ClientID'])
```

```python
df_temp.info()
```

```python
df_temp.head(5)
```

```python
# Merging customers and services - using difents columns name to merge
df_churn_temp = df_temp.merge(df_contracts, left_on=['ClientID'], right_on=['customerID'])
```

```python
df_churn_temp.info()
```

```python
# Merge all dataframes - With diferents columns name 
df_churn = df_customers.merge(
  df_services, on=['ClientID']).merge(df_contracts, left_on=['ClientID'], right_on=['customerID']
)
```

```python
df_churn.info()
```

#### Removing columns

```python
df_churn.drop(['customerID'], axis=1, inplace=True)
```

```python
df_churn.info()
```

### Missing values detection

```python
# All columns of a dataframe
df_churn.isna().sum()
```

```python
# A columns of a dataframe
df_churn.TotalCharges.isna().sum()
```

```python
# How many rows have at least one missing value
df_churn[df_churn.isna().any(axis=1)]
```

```python
# How many columns have at least one missing value
df_churn.isna().any(axis=0).sum()
```

#### Dealing with missing values

```python
# Remove the column with missing value
df_churn.drop(columns=['TotalCharges'], axis=1)
```

```python
# Remove columns with missing values
df_churn.dropna(axis=1)
```

```python
# Remove columns where ALL values are missing
df_churn.dropna(axis=1, how='all')
```

#### Input missing values

```python
# Fill all values of the dataframe with 0
df_churn.fillna(0)
```

```python
# Fill default values depending on 
df_churn.fillna(value={'TotalCharges': 0, 'Genero': 'Não declarado'})
```

```python
# Fill all missing values with a mean
mean_totalCharges = df_churn.TotalCharges.mean()
mean_totalCharges
```

```python
df_churn.fillna(value={'TotalCharges': mean_totalCharges })
```
